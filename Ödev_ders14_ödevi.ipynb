```python
#en yüksek puan ort. ile samp olan takım:mac sayısı da onemli,orn 90 puan 30 mac:1 mac ort 3 oldu
sampiyon["ort_puan"]=sampiyon["puan_duzelmis"]/sampiyon["oyun"]

maxortpuan=sampiyon["ort_puan"].max()
sampiyon[sampiyon["ort_puan"]==maxortpuan]
```

```python
#en az puan ort. ile samp olan takım

minortpuan=sampiyon["ort_puan"].min()
sampiyon[sampiyon["ort_puan"]==minortpuan]
```
